```python
xml_data = response.text

# XML 파싱
root = ET.fromstring(xml_data)

# 'items' 태그 안의 모든 'item' 태그 찾기
items = root.findall('.//item')

# 각 항목에 대해 데이터 추출
data = []
for item in items:
    record = {
        'aptDong': item.find('aptDong').text,
        'aptNm': item.find('aptNm').text,
        'aptSeq': item.find('aptSeq').text,
        'bonbun': item.find('bonbun').text,
        'bubun': item.find('bubun').text,
        'buildYear': item.find('buildYear').text,
        'buyerGbn': item.find('buyerGbn').text,
        'cdealDay': item.find('cdealDay').text,
        'cdealType': item.find('cdealType').text,
        'dealAmount': item.find('dealAmount').text,
        'dealDay': item.find('dealDay').text,
        'dealMonth': item.find('dealMonth').text,
        'dealYear': item.find('dealYear').text,
        'dealingGbn': item.find('dealingGbn').text,
        'estateAgentSggNm': item.find('estateAgentSggNm').text,
        'excluUseAr': item.find('excluUseAr').text,
        'floor': item.find('floor').text,
        'jibun': item.find('jibun').text,
        'landCd': item.find('landCd').text,
        'landLeaseholdGbn': item.find('landLeaseholdGbn').text,
        'rgstDate': item.find('rgstDate').text,
    }
    data.append(record)

# 결과 열 생략없이 출력하도록
pd.set_option('display.max_columns', None)
    
# DataFrame으로 변환
df = pd.DataFrame(data)


# 결과 출력
display(df)
```

```python
import pandas as pd
import requests
from time import sleep

# DataFrame을 저장할 리스트
all_data = []

# 2000년부터 2024년까지 월별 데이터를 요청
for year in range(2000, 2025):
    for month in range(1, 13):
        # 연월 (YYYYMM) 형식으로 DEAL_YMD 생성
        deal_ymd = f"{year}{month:02d}"

        # 요청 파라미터 설정
        params = {
            "serviceKey": de_key, 
            "LAWD_CD": 11110,               # 지역코드
            "DEAL_YMD": deal_ymd,           # 계약월
            "pageNo": 1,                    # 페이지 번호
            "numOfRows": 1000,              # 한 페이지에 최대 1000개까지 불러오기
        }

        # API 요청
        response = requests.get(base_url, params=params)

        # 요청이 성공한 경우 데이터 처리
        if response.status_code == 200:
            print(f"요청 성공: {deal_ymd}")

            # XML 파싱 후 데이터 추출
            root = ET.fromstring(response.text)
            items = root.findall('.//item')

            # 데이터를 리스트에 추가
            for item in items:
                record = {
                    'aptDong': item.find('aptDong').text,
                    'aptNm': item.find('aptNm').text,
                    'aptSeq': item.find('aptSeq').text,
                    'bonbun': item.find('bonbun').text,
                    'bubun': item.find('bubun').text,
                    'buildYear': item.find('buildYear').text,
                    'buyerGbn': item.find('buyerGbn').text,
                    'cdealDay': item.find('cdealDay').text,
                    'cdealType': item.find('cdealType').text,
                    'dealAmount': item.find('dealAmount').text,
                    'dealDay': item.find('dealDay').text,
                    'dealMonth': item.find('dealMonth').text,
                    'dealYear': item.find('dealYear').text,
                    'dealingGbn': item.find('dealingGbn').text,
                    'estateAgentSggNm': item.find('estateAgentSggNm').text,
                    'excluUseAr': item.find('excluUseAr').text,
                    'floor': item.find('floor').text,
                    'jibun': item.find('jibun').text,
                    'landCd': item.find('landCd').text,
                    'landLeaseholdGbn': item.find('landLeaseholdGbn').text,
                    'rgstDate': item.find('rgstDate').text,
                }
                all_data.append(record)

            # 요청 간 간격을 두어 서버에 부담을 주지 않도록 함
            sleep(0.5)  # 0.5초 대기

        else:
            print(f"요청 실패: {deal_ymd}, 상태 코드: {response.status_code}")

# 수집된 데이터를 pandas DataFrame으로 변환
df = pd.DataFrame(all_data)

# 결과 출력
print(df)

# 결과를 CSV로 저장 (필요시)
df.to_csv('apt_trade_data.csv', index=False)
```
